### **EXTRACTING INFORMATION FROM HTML FILE:**

1. Parsing the specified directory 'ddrFinalProject', with iterating over all the files and opening/ reading if html file.
2. os.path.join() is used to specify the path of file in folder, while os.path.basname() specifies filename
3. os.path.isfile() is used to check if the filepath does specify a file or not
4. The file is then open with 'r' mode. 'r' is for reading mode.
5. Selecting the informations using select_one(), find_all(), select() methods and using .text to extract the texts field. .strip() is used to remove unwanted trailing or leading whitespaces.
6. The page is iterated with finding 'h3' which containts all the title in *<a>* anchor element. Iterating over each of the title, the information for single medicines is fetched.
7. If the item exists, title and url are selected. Other information are iterated by fetching the detailed page by clicking on the url. This page gives price, variant, description, image url link.
8. re.match(r"healthcare-(\w+)-page\d+\.html", filename), is used as a regex function to extract category from the file saved. While pushing files, we have saved it in this format. The middle word tells us about the tab. It specifies healthcare- followed by words followed by page(decimal number).html.
9. The data obtained is then stored in a dictionary which is then inserted into the list. This is done to store the data in a database. 

```python
headers = {'User-Agent': 'Mozilla/5.0'}

# Mentioning the directory name
directory = 'ddrFinalProject'

# Creating a list to store all the data in the form of dictionaries
productDataList = []

# Iterating for all files in directory
for file in os.listdir(directory):
    # Joining the path with filename for accesing
    filepath = os.path.join(directory, file)
    # Printing the filename
    filename = os.path.basename(filepath)
    print(filename)
    # Checking if the file is html and if the filepath specifies a file type or not
    if os.path.isfile(filepath) and file.endswith(".html"):
        with open(filepath, 'r', encoding='utf-8') as file:
            # Reading the file and parsing the html content
            html = file.read()
            #print(html)
            soup = BeautifulSoup(html, 'html.parser')
            # Extracting all 'h3' items
            products = soup.find_all('h3') 
            # Initialising count
            count = 0
            # Iterating over each item
            for item in products:
                # Checking if item exists
                if item.a:
                    # Printing information
                    title = item.a.text.strip()
                    print('Count:', count)
                    print()
                    print('Title:', title)
                    count = count+1
                    url = item.a['href']
                    print("Link:", url)
                    # fetching url for detailed page, adding time.sleep() to mimic human behaviour
                    time.sleep(7)
                    page = requests.get(url, headers=headers)
                    soup = BeautifulSoup(page.content, 'html.parser')
                    #print(soup.prettify())
                    # Fetching the details, adding if condition to avoid exceptions
                    productId = soup.find('span', itemprop='sku')
                    if productId:
                        productId = productId.text.strip()
                        print("ProductId:", productId)
                    price = soup.find('span', itemprop='price')
                    if price:
                        price = price.text.strip()
                        print("Price:", price)
                    # Using regex function to extract the word in middle 
                    match = re.match(r"healthcare-(\w+)-page\d+\.html", filename)
                    if match:
                        # Fetching the first word in ()
                        category = match.group(1)
                        print("Category:", category)
                    variant = soup.find('h3', itemprop='brand')
                    if variant:
                        variantName = variant.find('a').text.strip()
                        print("Variant:", variantName)
                    availability = soup.find('h5') 
                    if availability:
                        availability = availability.text.strip()
                        print('Availability:', availability)
                    description = soup.find('h4', string='Description:')
                    if description:
                        # Extracting by specifying next sibling
                        para = description.find_next_sibling('p')
                        if para:
                            print("Description:", para.text.strip())
                    img = soup.find('img')
                    if img:
                        print("Image:", img['src'])

                    # Storing all the information in dictionary, later to be used for inserting in DB

                    productData = {
                        'title': title,
                        'url': url,
                        'productId': productId,
                        'price': price,
                        'category': category if category else None,
                        'variant': variantName if variantName else None,
                        'availability': availability if availability else None,
                        'description': description if description else None,
                        'image': img if img else None
                    }

                    # Removing keys with None values from the dictionary
                    productData = {key: value for key, value in productData.items() if value is not None}

                    # Adding the data dictionaries to the list
                    productDataList.append(productData)
 
```

### **STORING THE DICTIONARY IN HTML FILE:**

1. This is done to ensure minimum data loss.
2. The starting of file is created by attaching the opening tags.
3. The products in the stored list is iterated and the information is then stored with format. Each product html is stored in main html.
4. The file is then closed with all the necessary tags, and is stored in the same directory.

```python
# Making the file by attaching the starting of html file
htmlFile = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
</head>
<body>
"""

# Iterating over the products in product list of dictionaries, creating div product to save it
for product in productDataList:
    productFile = f"""
    <div class="product">
        <h2>{product.get('title', '')}</h2>
        <img src="{product.get('image', '')}" alt="{product.get('title', '')}" style="width:100px;height:auto;">
        <p>Price: {product.get('price', '')}</p>
        <p>ID: {product.get('productId', '')}</p>
        <p>Category: {product.get('category', '')}</p>
        <p>Variant: {product.get('variant', '')}</p>
        <p>Availability: {product.get('availability', '')}</p>
        <p>Description: {product.get('description', '')}</p>
        <a href="{product.get('url', '#')}">Product Page</a>
    </div>
    <hr>
    """
    htmlFile += productFile
    # Appending product info to html file info

# Closing the tags
htmlFile += """
</body>
</html>
"""

# Writing it into the file 
# Mentioning the directory name
directory = 'ddrFinalProject'
file = "product_list.html"
filepath = os.path.join(directory, file)
with open(filepath, "w", encoding="utf-8") as file:
    file.write(htmlFile)

print(f"HTML file '{os.path.basename(filepath)}' has been created.")
```

### INSERTING THE DATA

1. Converting the html file tag structure into urls or string value. This is done to ensure bson compatibility.
2. The dictionary is copied to ensure no change in original data
3. The elements value is checked for <tag>, if it has img name in it, the key is saved as a url. Otherwise, value is converted into string.

```python
# Copying the dictionary
from copy import deepcopy
ProductDataListUpdated = deepcopy(productDataList)

# Iterating for reach product
for product in ProductDataListUpdated:
    # Iterating over each key value pair
    for key, value in product.items():
        # Checking if element is Tag, and then if it is img
        if isinstance(value, Tag):
            # src for images
            if value.name == 'img':
                product[key] = value['src']
            # If it is not image the value is converted into string, otherwise set as a url
            else:
                product[key] = str(value)
```

```python
# Insert data to MongoDB

collection.insert_many(productDataList)
print("Data saved to MongoDB.")
```

### INSIGHT 1: AVERAGE PRICE OF ALL THE PRODUCTS
----------------------------------------------------

The average price for all results is calulated using the aggregate function. The result is grouped by to get a singular result. toDecimal is specified, before converting price to average for avoiding any exceptions. Next function fetches the first element. Since, the output consists of only one element, next() can be used.

**INFERENCE:** The average price turns out to be $14.18. However, this could vary depending on the quantity of medicines, also on they type such as strip, bottles etc.

```python
# Fetching result after aggregating
result = db.products.aggregate([
    {"$group": {
        "_id": None,
        "averagePrice": {"$avg": {"$toDecimal": "$price"}}
    }}
])

avgeragePrice = next(result, {}).get('averagePrice', 'N/A')
avgeragePrice = str(avgeragePrice)
avgeragePriceRounded = round(float(avgeragePrice), 2)
print(f"Average Price: {avgeragePriceRounded}")
```

To deep dive into how it is seggregated within the categories, average price for each of them is checked. id is specified to which column we want to group. The todecimal converts it to decimal for computing average.

**INFERENCE:** Compared to the average price of all products, allergy medicines have a higher average price. Meanwhile, digestive medicines suggests a lower average price.  

```python
# Find the average price for all categories
result = db.products.aggregate([
    {"$group": {
        "_id": "$category", # grouping by category
        "averagePrice": {"$avg": {"$toDecimal": "$price"}}
    }}
])

# Printing category and price
for data in result:
    avgeragePrice = str(data['averagePrice'])
    print(f"Category: {data['_id']}, Average Price: {round(float(avgeragePrice), 2)}")
```

### INSIGHT 3: CHEAPEST PRODUCT
------------------------------------

After finding the average, we wanted to see the cheapest product on OTC business. This is done by taking the price and sorting in ascending order. The first output is taken for the lowest price.

**INTERESTING FIND:** According to MedlinePlus.gov, nonprescription ibuprofen is utilized to lower fever and alleviate minor aches and pains stemming from various conditions including headaches, muscle aches, arthritis, menstrual periods, the common cold, toothaches, and backaches. The broad application and affordability of ibuprofen may lead to misuse and abuse.
```python
# Finding the cheapest product
cheapestProduct = db.products.aggregate([
    {"$match": {"price": {"$exists": True}}}, # Checking if price exists
    {"$addFields": {"numeric_price": {"$toDecimal": "$price"}}}, 
    {"$sort": {"numeric_price": 1}}, # sorting in ascending
    {"$limit": 1} # getting the first value
])

for product in cheapestProduct:
    print(f"Product {product['title']} has the lowest price of {product['numeric_price']} in category - {product.get('category')}, it also has a variant of '{product.get('variant')}'")
```

### INSIGHT 4: EXPENSIVE PRODUCT

Similarly, the last element is taken as the greatest value.

```python
# Finding the most expensive product
ExpensiveProduct = db.products.aggregate([
    {"$match": {"price": {"$exists": True}}},
    {"$addFields": {"numeric_price": {"$toDecimal": "$price"}}}, 
    {"$sort": {"numeric_price": -1}},
    {"$limit": 1}  
])

for product in ExpensiveProduct:
    print(f"Product {product['title']} has the highest price of {product['numeric_price']} in category - {product.get('category')}, it also has a variant of '{product.get('variant')}'")
```

### INSIGHT 5: AVAILABILITIES AMONG CATEGORIES
-------------------------------------------

The analysis then focused on how the categories have availabilities across.

**RECOMMENDATION:** Cold category has the highest out-of-stock ratio. Given the prevalence and seasonal nature of colds, we recommend pharmacies optimize their inventory management.

```python
# Calculating availabilities 
availabilityCategory = db.products.aggregate([
    {
        "$group": {
            "_id": {
                "category": "$category",
                "availability": "$availability"
            },
            "count": {"$sum": 1}  # Storing the count
        }
    }
])

# Initialising a dictionary
stockCategory = {}

# Iterating over each group
for group in availabilityCategory:
    # Extracting the information
    category = group["_id"]["category"]
    availability = group["_id"]["availability"]
    count = group["count"]

    # Initialising for the category key
    if category not in stockCategory:
        stockCategory[category] = {"In Stock": 0, "Out of Stock": 0}

    # Incrementing the count
    if availability == "In Stock":
        stockCategory[category]["In Stock"] += count
    elif availability == "Out of Stock":
        stockCategory[category]["Out of Stock"] += count

# Printing the results
for category, counts in stockCategory.items():
    print(f"CATEGORY: {category}, In Stock: {counts['In Stock']}, Out of Stock: {counts['Out of Stock']}")
```

### INSIGHT 7: VARIANT ANALYSIS
-----------------------------------

Finally, it was checked how variants for the products vary. Variants are the medicines that could be taken in place if the recommended medicine is out of stock. The analysis was done by grouping the products on variant and checking the number attached to it.

**INFERENCE:** Zyrtec, Voltaren, Extra Strength Tylenol are the top 3 variants that have the most product numbers.

```python
# Grouping the products based on variants
variants = db.products.aggregate([
    {
        "$group": {
            "_id": "$variant",
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {"count": -1}  # Sorting on descending
    }
])

# Printing the results
for variant in variants:
    print(f"Variant: {variant['_id']}, Number of Products: {variant['count']}")
```
